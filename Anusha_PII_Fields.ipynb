| Field               | PII Type                          | Action                            | Justification                                                                                                               |
| ------------------- | --------------------------------- | --------------------------------- | --------------------------------------------------------------------------------------------------------------------------- |
| **full_name**       | Direct PII                        | **Drop or Pseudonymize**          | Directly identifies an individual. Not needed for research → safest to drop. If tracking is needed, replace with unique ID. |
| **email**           | Direct PII                        | **Drop**                          | Highly sensitive and uniquely identifying. No analytical value for most health studies.                                     |
| **date_of_birth**   | Indirect PII                      | **Mask (e.g., year only)**        | Can identify individuals when combined with other fields. Keep only age or birth year for analysis.                         |
| **zip_code**        | Indirect PII                      | **Mask (e.g., first 2–3 digits)** | Can re-identify individuals when combined with DOB/gender. Reduce granularity.                                              |
| **job_title**       | Indirect PII                      | **Keep or Generalize**            | Low risk alone, but can identify rare roles. Safer to group into categories (e.g., “Healthcare”, “IT”).                     |
| **diagnosis_notes** | Indirect PII (and sensitive data) | **Mask / De-identify**            | May contain embedded identifiers (names, locations). Use NLP or regex to remove personal references before sharing.         |
